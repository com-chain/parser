```python
def add_event(evt,shortname,currency):
    return [
        f'var evt_{shortname}_{evt} = contract_{shortname}'+'({}, {fromBlock: FBLOCK, toBlock: \'LBLOCK\'});',
        f'evt_{shortname}_{evt}.watch(function(error, result)'+
        ' {console.log(JSON.stringify('+
        f'add_Currency(result,"{currency}")))'+'});'
    ]

def add_currency(contract, name, shortname):
    result = [
        ' ',
        ' ',
        ' ',
        f'// {name}',
        f'const addr_{shortname} = "{contract}"',
        f'var contract_{shortname} = eth.contract(desc).at(addr_{shortname})'
    ]
    
    for evt in ['Transfer','TransferCredit','Pledge']:
        result = [*result,*add_event(evt,shortname,name)]
    return result
```
